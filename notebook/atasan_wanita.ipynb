```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import pandas as pd
import time
from selenium.webdriver.chrome.options import Options
import undetected_chromedriver as uc
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
driver = uc.Chrome()
```

```python
driver.get("https://www.tokopedia.com/p/fashion-wanita/atasan-wanita?ob=5&page=1")
```

```python
dct_data = {
    "Nama":[],
    # "Terjual":[],
    # "Toko":[],
    # "Daerah":[],
    # "Rating":[],
    "Harga":[],
    "Link":[]
}
```

```python
# EXAMPLE SCROLL AND NEXT PAGE

dct_class = {
    "css-20kt3o":"Nama"
    # ,"eLOomHl6J3IWAcdRU8M08A==":"Terjual"
    # ,"X6c-fdwuofj6zGvLKVUaNQ== -9tiTbQgmU1vCjykywQqvA== flip":"Toko"
    # ,"-9tiTbQgmU1vCjykywQqvA== flip":"Daerah"
    # ,"nBBbPk9MrELbIUbobepKbQ==":"Rating"
    ,"css-o5uqvq":"Harga"
    # ,"_8cR53N0JqdRc+mQCckhS0g== ":"Harga"
    ,"css-54k5sq":"Link"
}

for i in range(56):
    print(f'page {i}')
    time.sleep(3)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # WEBSCRAPPING try 3 times if content exists
    try_time = 3
    for tr in range(try_time):
        try:
            div_data = driver.find_element(By.XPATH, '//div[@data-testid="lstCL2ProductList"]')
            break
        except:
            time.sleep(2)
            try_time -= 1
    
    if try_time == 0:
        print("There is error after trying over 3 times")
        break

    #SPAN
    find_span = div_data.find_elements(By.CSS_SELECTOR, 'span')
    for span in find_span:
        if span.get_attribute('class') in dct_class:
            field = dct_class[span.get_attribute('class')]
            dct_data[field].append(span.text)
            
    #DIV
    find_div = div_data.find_elements(By.CSS_SELECTOR, 'div')
    for div in find_div:
        if div.get_attribute('class') in dct_class:
            field = dct_class[div.get_attribute('class')]
            dct_data[field].append(div.text)
            
    #DIV
    find_a = div_data.find_elements(By.CSS_SELECTOR, 'a')
    for a in find_a:
        if a.get_attribute('class') in dct_class:
            field = dct_class[a.get_attribute('class')]
            dct_data[field].append(a.get_attribute('href'))

    # Next Page
    element = driver.find_element(By.XPATH, '//*[@aria-label="Laman berikutnya"]')
    driver.execute_script("arguments[0].click();", element)
    # next_page = driver.find_element(By.XPATH, '//*[@aria-label="Laman berikutnya"]')
    # next_page.click()
```
