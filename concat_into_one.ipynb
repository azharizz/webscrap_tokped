```python
import pandas as pd
```

```python
l1 = pd.read_csv('aksesoris_pria.csv')
l2 = pd.read_csv('atasan_pria.csv')
l3 = pd.read_csv('batik_pria.csv')
l4 = pd.read_csv('celana_pria.csv')
l5 = pd.read_csv('dalaman_pria.csv')
l6 = pd.read_csv('outwear_pria.csv')
```

```python
# Concatenate DataFrames into a single DataFrame, each DataFrame forming a row
all_dfs = [l1, l2, l3, l4, l5, l6]
l_concat = pd.concat(all_dfs, axis=0, ignore_index=True)
```

```python
l_concat
```

```python
l_concat['gender'] = 'pria'
```

```python
l_concat
```

```python
w1 = pd.read_csv('aksesoris_wanita.csv')
w2 = pd.read_csv('atasan_wanita.csv')
w3 = pd.read_csv('batik_wanita.csv')
w4 = pd.read_csv('celana_wanita.csv')
w5 = pd.read_csv('dalaman_wanita.csv')
w6 = pd.read_csv('outwear_wanita.csv')
```

```python
all_dfs = [w1, w2, w3, w4, w5, w6]
w_concat = pd.concat(all_dfs, axis=0, ignore_index=True)
```

```python
w_concat
```

```python
w_concat['gender'] = 'wanita'
```

```python
w_concat
```

```python
lw_concat = pd.concat([l_concat, w_concat], axis=0, ignore_index=True)
```

```python
lw_concat = lw_concat.reset_index()
```

```python
lw_concat
```

```python
def clean_price(price):
    # Remove 'Rp' and dots, then convert to integer
    return int(price.replace('Rp', '').replace('.', ''))

# Apply the function to the price column
lw_concat['Harga'] = lw_concat['Harga'].apply(clean_price)
```

```python
lw_concat
```

```python
lw_concat.to_csv('all_clean_scrap.csv', index=False)
```
